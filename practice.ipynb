```python
def testing():
    word = 'George'
    acceptable_range = range(0,10)
    within_range = False
    while word.isdigit() == False or within_range == False:
        word = input('enter a valid digit from 0-10')
        
        ##digit check
        if word.isdigit() == False:
            print('error!!you need to insert a digit')
         ###range check
        if word.isdigit() == True:
            if int(word) in acceptable_range:
                within_range = True
            else:
                within_range = False
                print('sorry!! you are out of acceptable range.')
    return int(word)
testing()
```

```python
def gameplay_choice():
    choice = 'wrong'
    while choice not in ['Y','N']:
        choice = input('select either Y or N:')
        if choice not in ['Y','N']:
            print('oops!!,you need to select a valid option of Y or N')
    if choice == 'Y':
        return True
    else:
        return False
gameplay_choice()
    
```

```python
###init is called upon whenever we create an instance of a class.
class dog():
    species = 'mammal'# class object attribute
    
    def __init__(self,breed,name,spots):
        #MAttribute
        #my_breed is also the argument name we recieve into the function
        #we assign it to self.attribute_name i.e whatever attribute name we choose to use
        self.breed = breed
        self.name = name
        self.spots = spots #we should be expecting a boolean here for spots.

        ##operations/methods
    def bark(self,number):
        print(f"woof my name is {self.name} and number is {number}")
```

```python
class circle():
    pi = 3.14 #general class Object attribute
    def __init__(self,radius):
        self.radius = radius 

        #operations/methods which is the function call in object oriented programming
    def get_circumference(self):
        return self.radius * self.pi * 2
         
    
```

```python
##object oriented programming special mehods
class book():
    def __init__(self,title,author,page):
        self.title = title
        self.author = author
        self.page = page

    def __str__(self):
        return f'the title of the book is {self.title} and written by {self.author},it has {self.page} pages'
    def __len__(self):
        return self.page
    def __del__(self):
        print('you just deleted the book class')
         
```

```python
import math
class line:
    def __init__(self,cod_1,cod_2):
        self.cod_1 = cod_1
        self.cod_2 = cod_2
    def distance(self):
        x1, y1 = self.cod_1
        x2, y2 = self.cod_2
        # Apply the distance formula
        return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
        pass
    def slope(self):
        x1,y1 = self.cod_1
        x2,y2 = self.cod_2
        return (y2 - y1) / (x2 - x1)
        pass
        
```

```python
a= (2,5)
b= (6,1)
my_cod = line(a,b)
my_cod.distance()
```

```python
class bank_account:
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance
    def owner(self):
        return self.owner
    def balance(self):
        return self.balance
    def deposit(self,dep_amt):
        self.balance = self.balance + dep_amt
        print(f'{dep_amt} is added successfully')
    def withdrawal(self,wd_amt):
        if self.balance >= wd_amt:
            self.balance = self.balance - wd_amt
            print('you have withrawn successfully')
        else:
            print('sorry,you do not have enough funds')
    def __str__(self):
         return(f'The owner of the account is {self.owner} and the balance is {self.balance}')
```
