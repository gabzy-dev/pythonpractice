```python
#Write a program that prints the integers from 1 to 100. But for multiples of three print "Fizz" instead of the number, and for the multiples of five print "Buzz". For numbers which are multiples of both three and five print "FizzBuzz".


for numbers in range(1,101):
    if numbers % 5 == 0 and numbers % 3 == 0:
        print('fizzbuzz')
    elif numbers % 5 == 0:
        print('buzz')
    elif numbers % 3 == 0:
        print('fizz')
    else:
        print(numbers)
```

```python
work_hours = [('dare',39),('george',89),('jerry',699),('shaw',54)]

def work_time_func(work_hours):
    "finding the employee of the month"
    work_hours_current = 0
    employee_name = ''
    for employee,hours in work_hours:
        if hours > work_hours_current:
            work_hours_current = hours
            employee_name = employee
        else:
            pass
    return (employee_name,work_hours_current)
work_time_func(work_hours)
    

```

```python
def my_func(a):
    res = ' '
   
    for index,word in enumerate(a):
        if index % 2  == 0:
             res = res + word.upper()
        else:
            res = res + word.lower()
    return res
    
my_func('gorilla')
```

```python
def sum_integers(a,b):
    if (a+b) == 20:
        return True
    elif a or b == 20:
        return True
    else:
        return False
sum_integers(0,20)
```

```python
def capital_word(a):
    first_letter = a[:3]
    second_letter = a[3:]
    v = first_letter.capitalize()
    vv = second_letter.capitalize()
    return v + vv
capital_word('jejekekeke')
```

```python
def reverse(a):
    word = a.split()
    new_word = word[::-1]
    return new_word
reverse('i am home now')
```

```python
def three_char(n):
    result = ''
    for i in n:
        result = result + i*3
    return result
three_char('goat')
    
```

```python
####LAMBDA EXPRESSIONS!!!!!!!!!!!!!!!!!!!!!!!!!
def even(word):
    if len(word) % 2 == 0:
        return 'even_word'
    else:
        return word[0]

word = ['George','Dare','qudus']
list(map(even,word))
```
